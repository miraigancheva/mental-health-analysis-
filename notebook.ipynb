# Mental Health of International Students in Japan

## Background

A Japanese international university surveyed its students in 2018 to investigate whether studying abroad affects mental health outcomes. The study was reviewed and approved by multiple ethical and regulatory boards, and its findings were published in *Sustainability* (MDPI) in 2019.

> **Citation:** Nguyen, M. H., Le, T. T., & Meirmanov, S. (2019). Depression, Acculturative Stress, and Social Connectedness among International University Students in Japan: A Statistical Investigation. *Sustainability*, 11(3), 878. https://doi.org/10.3390/su11030878

The study found that **international students face a higher risk of mental health difficulties** than the general population. Two key predictors of depression were identified:

- **Social connectedness** — the degree to which a student feels a sense of belonging to a social group
- **Acculturative stress** — psychological stress associated with adapting to a new culture

This analysis replicates the core findings using SQL and further examines whether **length of stay** is a contributing factor to depression severity among international students.

## Dataset

The `students` table contains survey responses from both international (`Inter`) and domestic (`Dom`) students. The key columns used in this analysis are described below.

| Field Name      | Description                                                          |
| --------------- | -------------------------------------------------------------------- |
| `inter_dom`     | Student type: international (`Inter`) or domestic (`Dom`)            |
| `japanese_cate` | Japanese language proficiency (Low / Average / High)                 |
| `english_cate`  | English language proficiency (Low / Average / High)                  |
| `academic`      | Current academic level (undergraduate or graduate)                   |
| `age`           | Current age of student                                               |
| `stay`          | Current length of stay in years                                      |
| `todep`         | Total depression score — PHQ-9 test (higher = more severe)           |
| `tosc`          | Total social connectedness score — SCS test (higher = better)        |
| `toas`          | Total acculturative stress score — ASISS test (higher = more stress) |

## 2. Mental Health Scores by Length of Stay (International Students)

To assess whether time spent in the host country influences mental health, we filter for international students and group by length of stay. For each group we calculate:

- **`count_int`** — number of international students at that stay duration
- **`average_phq`** — mean depression score (PHQ-9); scores ≥ 10 indicate moderate-to-severe depression
- **`average_scs`** — mean social connectedness score (SCS); higher scores indicate a stronger sense of belonging
- **`average_as`** — mean acculturative stress score (ASISS); higher scores indicate greater cultural adjustment stress

Results are ordered by length of stay (descending) to observe trends over time.

```python
SELECT
    stay,
    COUNT(stay)          AS count_int,
    ROUND(AVG(todep), 2) AS average_phq,
    ROUND(AVG(tosc), 2)  AS average_scs,
    ROUND(AVG(toas), 2)  AS average_as
FROM students
WHERE inter_dom = 'Inter'
  AND stay IS NOT NULL
GROUP BY stay
ORDER BY stay DESC;
```

## 3. Summary of Findings

The results broadly support the conclusions of the original 2019 study:

- **Depression (PHQ-9):** Scores are elevated across all stay durations. Students who have been in Japan for 3–4 years show some of the highest average depression scores, suggesting that mental health challenges are not limited to the initial adjustment period.
- **Social connectedness (SCS):** Scores do not show a simple linear improvement over time, indicating that building a genuine sense of belonging remains difficult even after several years abroad.
- **Acculturative stress (ASISS):** Stress levels are consistently high across all groups, reinforcing the finding that cultural adjustment is a sustained challenge rather than a short-term one.

> **Note:** Sample sizes for stays of 5 years or more are very small (n ≤ 3), so averages for those groups should be interpreted with caution.
